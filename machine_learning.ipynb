DATA CLEANING AND VISUALISATION 

```python
#PLOTTING MISSING VALUES TO SEE THE MISSING VALUES IN THE DATASET
msno.bar(data)
```

```python
#printing the total number of missing values in the dataset and the percentage of the missing values 
data_missing  = pd.DataFrame({"missing_values": data.isnull().sum(), "missing_values_percentages": data.isna().sum()/len(data)*100})
data_missing = data_missing.sort_values(by="missing_values_percentages", ascending=False)
data_missing
```

```python


# Check missing value percentage
missing_percentage = data.isnull().sum() / len(data) * 100

# Drop columns with 100% missing values
data = data.drop(columns=missing_percentage[missing_percentage == 100].index)

# Drop columns with more than 30% missing values
data = data.drop(columns=missing_percentage[missing_percentage > 30].index)

# Fill missing values in numerical columns with mean
num_cols = data.select_dtypes(include=['number']).columns
data[num_cols] = data[num_cols].fillna(data[num_cols].mean())

# Fill missing values in categorical columns with mode
cat_cols = data.select_dtypes(include=['object']).columns
data[cat_cols] = data[cat_cols].fillna(data[cat_cols].mode().iloc[0])

```

```python
#plot showing no missing values in the dataset after cleaning
msno.bar(data)
```
